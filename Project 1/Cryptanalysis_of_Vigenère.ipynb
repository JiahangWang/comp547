```python
# Define the alphabet (A = 0, ..., Z = 25)
alphabet = {chr(i + ord('A')): i for i in range(26)}

def calculate_S_values(ciphertext, period=17):
    n = len(ciphertext)
    S_values = []

    # Loop over each starting position i from 1 to period
    for i in range(period):
        # Extract the sequence c_i, c_{i+17}, c_{i+2*17}, ..., considering only uppercase characters
        sequence = [ciphertext[j] for j in range(i, n, period) if ciphertext[j] in alphabet]

        # Count frequencies of each character in the sequence
        frequency_count = Counter(sequence)

        # Normalize frequencies: q_i is the relative frequency of each letter
        total_count = sum(frequency_count.values())
        q_i = {alphabet[char]: count / total_count for char, count in frequency_count.items()}

        # Calculate S_i = sum(q_i^2) for the current sequence
        S_i = sum((q_i.get(i, 0) ** 2) for i in range(26))  # Use numeric index 0 to 25

        # Store the result
        S_values.append(S_i)

    return S_values


# Calculate S values for the given ciphertext and period 17
S_values = calculate_S_values(cipher_text, period=17)
S_values
```

```python
# Re-plot the histogram and overlay the normal distribution curve with the mean value indicated

plt.figure(figsize=(10, 6))

# Generate the histogram of S_values
count, bins, ignored = plt.hist(S_values, bins=10, density=True, alpha=0.6, color='g')

# Fit a normal distribution to the data
mu, std = np.mean(S_values), np.std(S_values)
xmin, xmax = plt.xlim()
x = np.linspace(xmin, xmax, 100)
p = stats.norm.pdf(x, mu, std)

# Plot the normal distribution curve
plt.plot(x, p, 'k', linewidth=2)

# Plot the mean as a vertical line
plt.axvline(mu, color='r', linestyle='dashed', linewidth=2)
plt.text(mu + 0.01, max(p) * 0.9, f'Mean: {mu:.2f}', color='r')

# Titles and labels
plt.title('Histogram of S_i Values with Normal Distribution Fit and Mean')
plt.xlabel('S_i Value')
plt.ylabel('Density')

plt.show()
```

```python
# English letter frequencies, p_j for j=0 to 25 corresponding to A to Z
english_frequencies = np.array([
    0.08167, 0.01492, 0.02782, 0.04253, 0.12702,  # A - E
    0.02228, 0.02015, 0.06094, 0.06966, 0.00153,  # F - J
    0.00772, 0.04025, 0.02406, 0.06749, 0.07507,  # K - O
    0.01929, 0.00095, 0.05987, 0.06327, 0.09056,  # P - T
    0.02758, 0.00978, 0.02360, 0.00150, 0.01974,  # U - Y
    0.00074  # Z
])


# Alphabet to index dictionary
alphabet = {chr(i): i - ord('A') for i in range(ord('A'), ord('Z') + 1)}
index_to_char = {i: chr(i + ord('A')) for i in range(26)}

# Function to calculate frequency of each letter in a sequence
def calculate_frequencies(sequence):
    total_count = len(sequence)
    frequencies = np.zeros(26)

    # Count the frequency of each character that is in the alphabet
    counter = Counter([letter for letter in sequence if letter in alphabet])

    for letter, count in counter.items():
        frequencies[alphabet[letter]] = count / total_count

    return frequencies

# Split the ciphertext into sequences with step 17
sequences = []
for i in range(17):
    sequence = cipher_text[i::17]  # Take every 17th character starting at index i
    sequences.append(sequence)

# Now calculate S_i for each sequence and find k_i
k_i = []

for i in range(17):
    sequence = sequences[i]
    q_i_j = calculate_frequencies(sequence)

    max_S_i = -float('inf')
    best_k = 0

    # For each k, calculate S_i and check the condition
    for k in range(26):
        S_i = np.sum(english_frequencies * np.roll(q_i_j, -k))  # q_{i,j+k}

        if S_i >= 0.06:
            best_k = k
            break  # We found a valid k, no need to check further

    k_i.append(best_k)

# Convert k_i values to corresponding letters
key_letters = ''.join([index_to_char[k] for k in k_i])

# Output the results
print(f"The key is: {key_letters}")
```

```python
# Function to decrypt Vigenère cipher
def vigenere_decrypt(ciphertext, key):
    decrypted_text = []
    for i in range(len(ciphertext)):
        # Convert characters to 0-25 range (A=0, B=1, ..., Z=25)
        cipher_char = ord(ciphertext[i]) - ord('A')
        key_char = ord(key[i % len(key)]) - ord('A')
        # Apply the decryption formula
        plain_char = (cipher_char - key_char + 26) % 26
        # Convert back to letter and append to the decrypted text
        decrypted_text.append(chr(plain_char + ord('A')))
    return ''.join(decrypted_text)


key = "QAOBKGLTCDVRRTZVL"

# Decrypt the ciphertext
plain_text = vigenere_decrypt(cipher_text, key)
plain_text
```
